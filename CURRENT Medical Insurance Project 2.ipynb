```python
#create list to store insurance data
insurance_data = []

#import the csv module
import csv

# open the csv file and append to the insurance_data list
with open ('insurance.csv') as insurance_csv:
    insurance_reader = csv.DictReader(insurance_csv, delimiter = ',')
    for i in insurance_reader:
        insurance_data.append(i)
    
```

```python
#method that counts the number of male and female patiens and the total number of patients overall
# and returns the percentage of each gender 
def sex_count (data_file):
    male_count = 0
    female_count = 0
    for record in data_file:
        if record['sex'] == 'male':
            male_count +=1
        elif record['sex'] == 'female':
            female_count += 1
    total_patients = male_count + female_count
    #get percentage of male and female patients compared to the total file
    male_percent = round(((male_count / total_patients)*100),2)
    female_percent = round(((female_count / total_patients)*100),2)
    print('COUNT OF PATIENTS, SPLIT BY GENDER:')
    print('Total Patients : {}'.format(total_patients))
    print('Male Patients  : {}, which is {}% of total patients'.format(male_count, male_percent))
    print('Female Patients: {}, which is {}% of total patients'.format(female_count, female_percent))
    print('')
    print('The data is roughly evenly split by gender')

    
    
```

```python
# call the funtion passing in the insurance_data file
print(sex_count (insurance_data))
```

```python
# calculate average insruance costs by gender
def average_insurance_costs_gender (data_file):
        male_charges = 0
        female_charges = 0
        male_count = 0
        female_count =0
        for record in data_file:
            if record['sex'] == 'male':
                male_charges += float(record['charges'])
                male_count += 1
            elif record['sex'] == 'female':
                female_charges += float(record['charges'])
                female_count += 1
        total_charges = male_charges + female_charges
        avg_insurance_cost = round((male_charges + female_charges) / (male_count + female_count),2)
        avg_insurance_cost_male = round((male_charges / male_count),2)
        avg_insurance_cost_female = round((female_charges / female_count),2)
        difference = avg_insurance_cost_male - avg_insurance_cost_female
        #return(male_charges, female_charges, total_charges)
        print('AVERAGE ANNUAL INSURANCE COSTS:')
        print('Average Insurance Cost       : ${}.'.format(avg_insurance_cost))
        print('')
        print('Male Average Insurance Cost  : ${}'.format(avg_insurance_cost_male))
        print('Female Average Insurance Cost: ${}'.format(avg_insurance_cost_female))
        print('')
        print('Difference                   : ${}'.format(difference))
        print ('Therefore, on average ,males pay significantly more for health insurance than females.')
```

```python
#call the function, passing in the insurance_Data file
print (average_insurance_costs_gender(insurance_data))
```

```python
# next steps are to work out the average ages of patients, both by gender and overall

def average_age (data_file):
    male_age_sum = 0
    male_count = 0
    female_age_sum =0
    female_count = 0
    for record in data_file:
        if record['sex'] == 'male':
            male_age_sum += float(record['age'])
            male_count += 1
        elif record['sex'] == 'female':
            female_age_sum += float(record['age'])
            female_count += 1
    age_sum = male_age_sum + female_age_sum
    total_count = male_count + female_count
    male_avg_age = round((male_age_sum / male_count),2)
    female_avg_age = round((female_age_sum / female_count),2)
    avg_age = round ((age_sum / total_count),2)
    diff_age =round((female_avg_age - male_avg_age),2)
    print('AVERAGE AGE OF PATIENTS:')
    print('Overall       : {}'.format(avg_age))
    print('')
    print('Male avg age  : {}'.format(male_avg_age))
    print('Female avg age: {}'.format(female_avg_age))
    print('')
    print('Therfore, on average, female patients are {} years older than males.'.format(diff_age))
    print('')
    print('Next, look at avg insurance charges by age  bracket to examine changes by age.')
    print('HOWEVER, we cannot isolate other factors such as smoker status or bmi from the analysis.')

            
            
    
    
```

```python
# then look at average insurance cost by age bracket to see if medical insurance costs increase with age
# this will help us to see if male or female costs could be higher due to age

def age_bands (data_file):
    # create age band varaibles and set to zero
    total_18_25 = 0
    total_26_35 = 0
    total_36_45 = 0
    total_46_55 = 0
    total_56_65 = 0
    total_over_65 = 0
    count_18_25 = 0
    count_26_35 = 0
    count_36_45 = 0
    count_46_55 = 0
    count_56_65 = 0
    count_over_65 = 0
    
    for record in data_file:
        if float(record['age']) >= 18 and float(record['age']) <= 25:
            total_18_25 += float(record['charges'])
            count_18_25 += 1
        elif float(record['age']) >= 26 and float(record['age']) <= 35:
            total_26_35 += float(record['charges'])
            count_26_35 += 1
        elif float(record['age']) >= 36 and float(record['age']) <= 45:
            total_36_45 += float(record['charges'])
            count_36_45 += 1
        elif float(record['age']) >= 46 and float(record['age']) <= 55:
            total_46_55 += float(record['charges'])
            count_46_55 += 1
        elif float(record['age']) >= 56 and float(record['age']) <= 65:
            total_56_65 += float(record['charges'])
            count_56_65 += 1
        elif float(record['age']) >= 66:
            total_over_65 += float(record['charges'])
            count_over_65 += 1
    avg_18_25 = round((total_18_25 / count_18_25),2)
    avg_26_35 = round((total_26_35 / count_18_25),2)
    avg_36_45 = round((total_36_45 / count_18_25),2)
    avg_46_55 = round((total_46_55 / count_18_25),2)
    avg_56_65 = round((total_56_65 / count_18_25),2)
    avg_over_65 = round((total_over_65 / count_18_25),2)
   # return (total_18_25, total_26_35, total_36_45, total_46_55, total_56_65,total_over_65)
    print('AVERAGE INSURANCE COSTS BY AGE BRACKET:')
    print('18-25: ${}'.format(avg_18_25))
    print('26-35: ${}'.format(avg_26_35))
    print('36-45: ${}'.format(avg_36_45))
    print('46-55: ${}'.format(avg_46_55))
    print('56-65: ${}'.format(avg_56_65))
    print('over 65s: ${}'.format(avg_over_65))
    print('')
    print('Medical insurance costs generally increase with age.')
    print('Patients in the 18-25 and 26-35 age brackets pay significantly less in annual medical insurance bills than older patients.')
    print('However, the 46-55 age group spends more on annual medical insurance than the 56-65s.')
    print('WHY IS THIS?')
        
```

```python
#Look at average insurance costs by gender and age bracket
def avg_insurance_costs_gender_age(data_file):
    
# create age band varaibles and set to zero
    male_charges_18_25 = 0
    male_charges_26_35 = 0
    male_charges_36_45 = 0
    male_charges_46_55 = 0
    male_charges_56_65 = 0
    male_count_18_25 = 0
    male_count_26_35 = 0
    male_count_36_45 = 0
    male_count_46_55 = 0
    male_count_56_65 = 0
    female_charges_18_25 = 0
    female_charges_26_35 = 0
    female_charges_36_45 = 0
    female_charges_46_55 = 0
    female_charges_56_65 = 0
    female_count_18_25 = 0
    female_count_26_35 = 0
    female_count_36_45 = 0
    female_count_46_55 = 0
    female_count_56_65 = 0
    
    for record in data_file:
        if float(record['age']) >= 18 and float(record['age']) <= 25 and record['sex'] == 'male':
            male_charges_18_25 += float(record['charges'])
            male_count_18_25 += 1
        elif float(record['age']) >= 26 and float(record['age']) <= 35 and record['sex'] == 'male':
            male_charges_26_35 += float(record['charges'])
            male_count_26_35 += 1
        elif float(record['age']) >= 36 and float(record['age']) <= 45 and record['sex'] == 'male':
            male_charges_36_45 += float(record['charges'])
            male_count_36_45 += 1
        elif float(record['age']) >= 46 and float(record['age']) <= 55  and record['sex'] == 'male':
            male_charges_46_55 += float(record['charges'])
            male_count_46_55 += 1
        elif float(record['age']) >= 56 and float(record['age']) <= 65 and record['sex'] == 'male':
            male_charges_56_65 += float(record['charges'])
            male_count_56_65 += 1
        if float(record['age']) >= 18 and float(record['age']) <= 25 and record['sex'] == 'female':
            female_charges_18_25 += float(record['charges'])
            female_count_18_25 += 1
        elif float(record['age']) >= 26 and float(record['age']) <= 35 and record['sex'] == 'female':
            female_charges_26_35 += float(record['charges'])
            female_count_26_35 += 1
        elif float(record['age']) >= 36 and float(record['age']) <= 45 and record['sex'] == 'female':
            female_charges_36_45 += float(record['charges'])
            female_count_36_45 += 1
        elif float(record['age']) >= 46 and float(record['age']) <= 55  and record['sex'] == 'female':
            female_charges_46_55 += float(record['charges'])
            female_count_46_55 += 1
        elif float(record['age']) >= 56 and float(record['age']) <= 65 and record['sex'] == 'female':
            female_charges_56_65 += float(record['charges'])
            female_count_56_65 += 1
    male_avg_charges_18_25 = round((male_charges_18_25 / male_count_18_25),2)
    male_avg_charges_26_35 = round((male_charges_26_35 / male_count_26_35),2)
    male_avg_charges_36_45 = round((male_charges_36_45 / male_count_36_45),2)
    male_avg_charges_46_55 = round((male_charges_46_55 / male_count_46_55),2)
    male_avg_charges_56_65 = round((male_charges_56_65 / male_count_56_65),2)
    female_avg_charges_18_25 = round((female_charges_18_25 / female_count_18_25),2)
    female_avg_charges_26_35 = round((female_charges_26_35 / female_count_26_35),2)
    female_avg_charges_36_45 = round((female_charges_36_45 / female_count_36_45),2)
    female_avg_charges_46_55 = round((female_charges_46_55 / female_count_46_55),2)
    female_avg_charges_56_65 = round((female_charges_56_65 / female_count_56_65),2)
    diff_avg_charges_18_25 =  round((male_avg_charges_18_25 - female_avg_charges_18_25),2) 
    diff_avg_charges_26_35 =  round((male_avg_charges_26_35 - female_avg_charges_26_35),2)
    diff_avg_charges_36_45 =  round((male_avg_charges_36_45 - female_avg_charges_36_45),2)
    diff_avg_charges_46_55 =  round((male_avg_charges_46_55 - female_avg_charges_46_55),2)
    diff_avg_charges_56_65 =  round((male_avg_charges_56_65 - female_avg_charges_56_65),2)
    print('Average Insurance costs by sex:')
    print('18-25: Males: ${}    Females: ${}   Difference: ${}'.format(male_avg_charges_18_25, female_avg_charges_18_25, diff_avg_charges_18_25))
    print('26-35: Males: ${}   Females: ${}    Difference: ${}'.format(male_avg_charges_26_35, female_avg_charges_26_35, diff_avg_charges_26_35))
    print('36-45: Males: ${}   Females: ${}    Difference: ${}'.format(male_avg_charges_36_45, female_avg_charges_36_45, diff_avg_charges_36_45))
    print('46-55: Males: ${}    Females: ${}   Difference: ${}'.format(male_avg_charges_46_55, female_avg_charges_46_55, diff_avg_charges_46_55))
    print('56-65: Males: ${}   Females: ${}   Difference: ${}'.format(male_avg_charges_56_65, female_avg_charges_56_65, diff_avg_charges_56_65))
    print('')
    print('Male insurance charges are higher in all age brackets, with the greatest difference between the ages of 26 and 45')
    print('The least difference between male and female charges is seen in the 18-25 group')
    print('')
    print('Therefore it would seem that male gender on its own is a contributing factor to higher medical insurance costs')
    print('Are males or females ore likely to smoke? Could smoker status be driving up the insurance costs in males?')
    print('Or is male gender on its own, a risk factor identified in the calculations for medical insurance costs?')
    
```
